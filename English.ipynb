```python
import os
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text(text, exclude_words=None):
    if exclude_words is None:
        exclude_words = []

    # Tokenization
    words = word_tokenize(text)

    # Remove punctuation and convert to lowercase
    words = [word.lower() for word in words if word.isalpha()]

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    # Exclude specific words
    words = [word for word in words if word not in exclude_words]

    # Stemming
    stemmer = PorterStemmer()
    words = [stemmer.stem(word) for word in words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]

    # Join the processed words back into a sentence
    processed_text = ' '.join(words)

    return processed_text

def process_files_in_folder(folder_path):
    exclude_words = ['doc', 'docno', 'text']  # Add any additional words to exclude
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if os.path.isfile(file_path):
            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()
                print(f"\nDocument: {filename}")
                processed_content = preprocess_text(content, exclude_words=exclude_words)
                print("\nProcessed Content:\n", processed_content)

folder_path = '/content/business'
process_files_in_folder(folder_path)
```

```python
import os
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
from collections import defaultdict

# nltk.download('punkt')
# nltk.download('stopwords')
# nltk.download('wordnet')

def preprocess_text(text):
    words = word_tokenize(text.lower())
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word.isalpha() and word not in stop_words and word.isalpha()]
    stemmer = PorterStemmer()
    words = [stemmer.stem(word) for word in words]
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    return words

def build_inverted_index(folder_path):
    inverted_index = defaultdict(list)

    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        if os.path.isfile(file_path):
            with open(file_path, 'r', encoding='utf-8') as file:
                content = file.read()

                # Preprocess content
                preprocessed_content = preprocess_text(content)

                # Build inverted index
                terms = preprocessed_content
                for term in set(terms):
                    inverted_index[term].append((filename, terms.count(term)))

    return inverted_index

def print_all_posting_lists(inverted_index):
    print("\nPosting lists for all terms:")
    for term, posting_list in inverted_index.items():
        print(f"\nTerm: '{term}'")
        for doc_id, term_freq in posting_list:
            print(f"Document: {doc_id}, Term Frequency: {term_freq}")
        print(f"Document Frequency: {len(posting_list)}")

def boolean_retrieval(inverted_index, query):
    query_terms = preprocess_text(query)
    result_docs = set()

    for term in query_terms:
        if term in inverted_index:
            postings = set(doc_id for doc_id, _ in inverted_index[term])

            # Handle NOT operation
            if 'NOT' in query_terms and query_terms[query_terms.index('NOT') + 1] == term:
                result_docs.difference_update(postings)
            else:
                # Handle OR and AND operations
                if not result_docs:
                    result_docs.update(postings)
                elif 'AND' in query_terms:
                    result_docs.intersection_update(postings)
                elif 'OR' in query_terms:
                    result_docs.update(postings)

    return result_docs

folder_path = '/content/business'
inverted_index = build_inverted_index(folder_path)

print_all_posting_lists(inverted_index)

query_result = boolean_retrieval(inverted_index, 'easter AND airway OR concert NOT tenness')
print(f"\nBoolean Retrieval Result for query 'easter AND airway OR concert NOT tenness': {query_result}")
```
